La función preprocesamiento realiza la tokenización completa del texto que se le entregue, fue utilizada para procesar el corpus y para que no existan problemas al reconocer los tokens también se usará para parsear las consultas.

Como pueden observar el proceso es muy similar al que se realizó en la última sesión, pero se agregó una gran lista de caracteres a la lista de Stopwords.

```python
def preprocesamiento(texto, es_query = True):
    stoplist = stopwords.words('english')
    listastop=["'","...","''","#","<",">","}","{","]","[","´","?","¡","!","¿","/","*",
                '"',"-","--",";",":","::","=",")","(","&","|","'ve","'t","'s",
                "'ll","n't",">.<","<3","._.","~","<br","</","/>","<<",">>","~~",
                ",","d","na","m","''","´´","!!!!!",'"(','",','.;','[*','tm','").',
               "l",'"&',').*','://','www','http','"...',"=[", '@', '``']
    if(es_query):
        listastop.append(".")
    nuevostoplist= stoplist+listastop
    tokens = nltk.word_tokenize(texto)
    filtered = [token for token in tokens if token not in nuevostoplist]
    stemmed = []
    stemmer = SnowballStemmer("english")
    for word in filtered:
        stemmed.append(stemmer.stem(word))
    return stemmed
```

CorpusParser y QueryParser son dos clases que permitirán mantener el corpus y las consultas en memoria principal. Además QueryParser realiza el preprocesamiento recién definido a las consultas, no así para CorpusParser, ya que este procedimiento toma demasiado tiempo como para realizarlo cada vez que se necesite cargar el corpus.

```python
class CorpusParser:
    def __init__(self, filename="corpus_docs.txt"):
        self.filename = filename
        self.regex = re.compile('^#\s*\d+')
        self.corpus = dict()
        
    #Parseo de corpus adaptado a mis WSJ92_stemmed
    def parse(self):#Pasarle los nombres.txt
        f = open(self.filename,'r')
        blobs = []
        for s in f.readlines():
            s = s.strip()
            blobs.append(s)
        f.close()
        for docid in blobs:#for id in nombres...
            f_stemmed = open(path2+docid,'r')
            text=''
            for line in f_stemmed.readlines():
                line = line.strip()+' '
                text+=line
            self.corpus[docid] = nltk.word_tokenize(text) #dict[id] = contenido en tokens 
            f_stemmed.close()

    def get_corpus(self):
        return self.corpus


class QueryParser:
    def __init__(self, filename="querys.txt"):
        self.filename = filename
        self.queries = []

    def parse(self):
        file_querys =open(self.filename,'r')
        for line in file_querys.readlines():
            line = line.strip().split()
            line = preprocesamiento(' '.join(line))
            #preprocesar
            
            self.queries.append(line)
        file_querys.close()

    def get_queries(self):
        return self.queries
```

InvertedIndex será la clase que se encargue de crear el indice invertido que el motor requiere para poder llevar a cabo la recuperación de documentos por cada consulta que se ejecute. Para esta experiencia la recuperación será a través de la disyunción de los documentos que contengan cada palabra de la consulta, es decir, se recuperarán todos los documentos que presenten por lo menos una palabra de la consulta.

DocumentLengthTable se encargará de mantener en memoria principal el largo de los documentos, esto es muy útil para calcular eficientemente las funciones de ranking.

```python
class InvertedIndex:
    def __init__(self):
        self.index = dict()

    def __contains__(self, item):
        return item in self.index

    def __getitem__(self, item):
        return self.index[item]

    def add(self, word, docid):
        if word in self.index:
            if docid in self.index[word]:
                self.index[word][docid] += 1
            else:
                self.index[word][docid] = 1
        else:
            d = dict()
            d[docid] = 1
            self.index[word] = d

    #frequency of word in document
    def get_document_frequency(self, word, docid):
        if word in self.index:
            if docid in self.index[word]:
                return self.index[word][docid]
            else:
                raise LookupError('%s not in document %s' % (str(word), str(docid)))
        else:
            raise LookupError('%s not in index' % str(word))

    #frequency of word in index, i.e. number of documents that contain word
    def get_index_frequency(self, word):
        if word in self.index:
            return len(self.index[word])
        else:
            raise LookupError('%s not in index' % word)


class DocumentLengthTable:

    def __init__(self):
        self.table = dict()

    def __len__(self):
        return len(self.table)

    def add(self, docid, length):
        self.table[docid] = length

    def get_length(self, docid):
        if docid in self.table:
            return self.table[docid]
        else:
            raise LookupError('%s not found in table' % str(docid))

    def get_average_length(self):
        sum = 0
        for length in list(self.table.values()):
            sum += length
        return float(sum) / float(len(self.table))


def build_data_structures(corpus):
    idx = InvertedIndex()
    dlt = DocumentLengthTable()
    for docid in list(corpus.keys()):
        #build inverted index
        for word in corpus[docid]:
            idx.add(str(word), str(docid)) #agrega palabras al indice invertido
            
        #build document length table
        length = len(corpus[str(docid)])
        dlt.add(docid, length)
    return idx, dlt
```

score_BM25 es la función que se encarga de calcular el puntaje de ranking para cada documento recuperado, como pueden ver todos los valores utilizados están dentro de los estándares vistos en el curso.

```python
from math import log

k1 = 1.2
k2 = 100
b = 0.75
R = 0.0

def score_BM25(n, f, qf, r, N, dl, avdl):
    K = compute_K(dl, avdl)
    first = log( ( (r + 0.5) / (R - r + 0.5) ) / ( (n - r + 0.5) / (N - n - R + r + 0.5)) )
    second = ((k1 + 1) * f) / (K + f)
    third = ((k2+1) * qf) / (k2 + qf)
    return first * second * third

def compute_K(dl, avdl):
    return k1 * ((1-b) + b * (float(dl)/float(avdl)) )
```

La clase Dist_semantica es la encargada de manejar los embeddings. Con ella se cargan a memoria principal los modelos y se utilizan para encontrar palabras y documentos cercanos a las consultas.

```python
class Dist_semantica:
  
    def __init__(self):
        self.similarity = dict()
        self.embeddings = dict()
        self.vec_dim = 0
        self.model = None
    
    def load_embeddings(self,filename):
        self.model = gensim.models.Word2Vec.load("word2vec.model")
        self.embeddings = self.model.wv
        self.vec_dim = len(self.embeddings["year"])
    
    def get_embeddings(self):
        return self.embeddings
    
    #Calcula el vector AWE de una lista de palabras (Consulta o documento)
    def calc_vector(self,terminos):
        largo=0
        vector_final = np.zeros(self.vec_dim)
        for termino in terminos:
            try:
                v_i = self.embeddings[termino]
                norma = 1.0/norm(v_i)
                v_i = norma*v_i
                vector_final = np.sum([vector_final, v_i], axis=0)
                largo+=1
            except:
                continue
            
        if(largo!=0):
            length_inv = 1.0/largo
            vector_final = length_inv*vector_final
        
        return vector_final, norm(vector_final)
    
    
    def similaridad_coseno(self,q,d):
        vq,norm_vq = self.calc_vector(q)
        vd,norm_vd = self.calc_vector(d)
        if(norm_vq==0.0 or norm_vd==0.0):
            return 0.0
        punto = np.dot(vq.T,vd)
        denominador = norm_vq*norm_vd
        return punto/denominador

    #Entrega los terminos unicos de todos los documentos en el conjunto docid
    def get_terminos(self,documentos):
        terminos = set()
        for doc in documentos:
            tmp = set(doc)
            terminos = terminos | tmp
        return list(terminos)
      
    #metodo de similitud por centroide
    def centroide(self,query, documentos, cant):
        sim_t = dict()
        q_cent = np.zeros(self.vec_dim)
        for qu in query:
            try:
                q = self.embeddings[qu]
                try:    
                    q_cent = np.sum(q,q_cent)
                except:
                    for i in range(self.vec_dim):
                        q_cent[i]+=q[i]
            except:
                continue

        terminos = self.get_terminos(documentos)
        #################################################
        for te in terminos:
            try:
                t = self.embeddings[te]
                if(te not in query):
                    norm_vq = norm(q_cent)
                    norm_vt = norm(t)
                    punto = np.dot(q_cent.T,t)
                    denominador = norm_vq*norm_vt
                    S_cent_t = np.exp(punto/denominador)
                    sim_t[te] = S_cent_t
            except:
                continue

        sim_ordenada = sorted(sim_t.items(), key=operator.itemgetter(1))
        sim_ordenada.reverse() 
        top = []
        for par in sim_ordenada:
            doc_par, numero = par  
            top.append(doc_par)
            if(len(top)==cant):
                break
        return top
  
  
  #Cant terminos más semejantes a termino
    def sim_terminos(self,termino, documentos, cant):
        terminos = self.get_terminos(documentos)
        sim_t = {}
        for t in terminos:
            try:
                t = self.embeddings[t]
                sim = np.cos(t,termino)
                sim_t[t] = sim
            except:
                break
      
        sim_ordenada = sorted(sim_t.items(), key=operator.itemgetter(1))
        sim_ordenada.reverse() 
        top = []
        for par in sim_ordenada:
            top.append(par)
            if(len(top)==cant):
                return top
            
##############################################################################
#Pueden agregar o reutilizar todas las funciones que necesiten 
```

Como pueden ver: Una consulta entra al sistema usando conjunción y este devuelve un ranking preliminar. Haciendo uso del top 10 de documentos (por default) se arma el conjunto T, que corresponde al conjunto de todos los términos que aparecen en algun documento del ranking. Luego uno a uno se compara el vector AWE de la consulta con el vector de la palabra y se genera el Top K de términos más cercanos a la consulta. Por último, se ingresa el Top K como una nueva consulta a un motor con disyunción y el ranking encontrado se pondera con el ranking original a través de una variable Alfa (1-Alfa para el original y Alfa para la expansión), entregandose el ranking final.

```python
class QueryProcessor:
    def __init__(self, queries, corpus, vectores="vectors.txt"):
        self.queries = queries
        self.corpus = corpus
        self.distsem = Dist_semantica()
        #self.distsem.parse_embeddings(filename=vectores)
        self.distsem.load_embeddings(filename=vectores)
        print("Embeddings Parseados")
        self.index, self.dlt = build_data_structures(corpus) #Construye el indice invertido
        print("Indice invertido Construido")


    def run_work(self,queries,exp,model,limite_docs,num_term_exp,alfa):
        results=[]
        if(not(exp)):
            for query in queries:
                print("Query ", str(query[0]))
                print(query)
                results.append(self.run_query_con(query[1:]))
        else:
            for query in queries:
                print("Query ", str(query[0]))
                print(query)
                results.append(self.run_query_final(query[1:],model,limite_docs,num_term_exp, alfa))
        return results

    #En un principio se definió esta funcion para manejar procesos multiples 
    #en la evaluacion de consultas, pero fue reducida y por eso puede parecer
    #redundante (Ya que su función solo es llamar a run_work)
    def run(self, exp = False, model='null', limite_docs=10, num_term_exp = 5,alfa=0.3):
        results = self.run_work(self.queries,exp,model,limite_docs,num_term_exp,alfa)
        return results
      
    #Recuperacion por disyuncion (BOW-OR)
    def run_query_dis(self, query):
        query_result = dict()
        #print(query)
        for term in query:
            if term in self.index:
                doc_dict = self.index[term] # retrieve index entry
                for docid, freq in doc_dict.items(): #for each document and its word frequency
                    score = score_BM25(n=len(doc_dict), f=freq, qf=1, r=0, N=len(self.dlt), 
                                       dl=self.dlt.get_length(docid), avdl=self.dlt.get_average_length())
                    if docid in query_result:
                        query_result[docid] += score
                    else:
                        query_result[docid] = score
        return query_result
    
    #Recuperacion por conjuncion (BOW-AND)
    def run_query_con(self, query):
        query_result = dict()
        docs_query =[]
        
        for term in query:
            if term in self.index:
                doc_list = set(self.index[term])
                docs_query.append(doc_list)
        if(len(docs_query)!=0):
            doc_set = docs_query[0]
            for d in docs_query:
                doc_set = doc_set & d
        else:
            doc_set = set([])
        
        for term in query:
            if term in self.index:
                doc_dict = self.index[term] # retrieve index entry
                for docid, freq in doc_dict.items(): #for each document and its word frequency
                    if docid in doc_set:
                        score = score_BM25(n=len(doc_dict), f=freq, qf=1, r=0, N=len(self.dlt), 
                                           dl=self.dlt.get_length(docid), avdl=self.dlt.get_average_length()) 
                        if docid in query_result:
                            query_result[docid] += score
                        else:
                            query_result[docid] = score
        return query_result
      
    def run_query_final(self, query, model, limite_docs,num_term_exp, alfa=0.3):
        preliminar_results = self.run_query_con(query)
        
        preliminar_results_orden = sorted(preliminar_results.items(), key=operator.itemgetter(1))
        preliminar_results_orden.reverse() 
        top = []
        for par in preliminar_results_orden:
            doc_par, numero = par
            doc = self.corpus[doc_par]
            top.append(doc)
            if(len(top)==limite_docs):
                break
        
        #Top contiene el top de documentos encontrados para la query
        if(model == "centroide"):
            extension_query = self.distsem.centroide(query,top,num_term_exp)
        ####################################################################
        #Esta es la parte que deben crear SU_FUNCION debe retornar N terminos
        #Que mas cercanos a la consulta
        elif(model == "cos_mul"):
            extension_query = self.distsem.SU_FUNCION()
        
        print(extension_query)
        final_results = self.run_query_dis(extension_query)
        for doc in list(final_results.keys()):
            if doc in list(preliminar_results.keys()):
                preliminar_results[doc] *= (1-alfa)
                preliminar_results[doc] += alfa*final_results[doc]
            else:
                preliminar_results[doc] = alfa*final_results[doc]
        
        return preliminar_results
```

Con la función guardar_resultados pueden almacenar los resultados que les entregue el motor de búsquedas: 

- results corresponde a los resultados del motor
- qids son las ids de las consultas, estas se manejan un poco más adelante en el código
- filename es el archivo destino en que se guardarán los resultados OJO: no sobre escriban sus resultados
- k es el largo del top ranking que desean guardar 



```python
def guardar_resultados(results, qids, filename, k=10, verbose=False):
    c_qid = 0
    with open(filename,"w") as file_resultados:
        for result in results:
            sorted_x = sorted(result.items(), key=operator.itemgetter(1))
            sorted_x.reverse()
            index = 0
            if(len(sorted_x)>k):
                for i in sorted_x[:k]:
                    tmp = (qids[c_qid], i[0], index, i[1])
                    el_string ='{:>1}\t{:>4}\t{:>2}\t{:>12}\tBM25'.format(*tmp)
                    file_resultados.write(el_string+'\n')
                    if(verbose):
                        print(el_string)
                    index += 1
            else:
                for i in sorted_x[:]:
                    tmp = (qids[c_qid], i[0], index, i[1])
                    el_string ='{:>1}\t{:>4}\t{:>2}\t{:>12}\tBM25'.format(*tmp)
                    file_resultados.write(el_string+'\n')
                    if(verbose):
                        print(el_string)
                    index += 1
            c_qid += 1
```

La función read_qrels encontrará las listas de relevancias para cada consulta que utilizaremos.

```python
def read_qrels(filename="qrels.txt"):
    rels = dict()
    with open(filename,'r') as f_qrels:
        for linea in f_qrels.readlines():
            linea = linea.strip().split()
            llave_r = linea[0]
            docid_r = linea[2]
            if(int(linea[-1]) == 1):
                if(llave_r in list(rels.keys())):
                    rels[llave_r].append(docid_r)
                else:
                    rels[llave_r] = [docid_r]
    return rels
```

En la siguiente celda se define una función que calculará la Mean Precision (MP), Average Recall hasta la posición k (AR@k) y el Normalized Discounted Cumulative Gain hasta la posición k (NDCG@k).

```python
def dcg_at_k(r, k, method=0):
    r = np.asfarray(r)[:k]
    if r.size:
        if method == 0:
            return r[0] + np.sum(r[1:] / np.log2(np.arange(2, r.size + 1)))
        elif method == 1:
            return np.sum(r / np.log2(np.arange(2, r.size + 2)))
        else:
            raise ValueError('method must be 0 or 1.')
    return 0.

def ndcg_at_k(r, k, method=0):
    dcg_max = dcg_at_k(sorted(r, reverse=True), k, method)
    if not dcg_max:
        return 0.
    return dcg_at_k(r, k, method) / dcg_max

def calcular_metricas(files, rels, k = 10, verbose=False): 
    with open(files,'r') as f_resultados:
        resultados = dict()
        for line in f_resultados.readlines():
            line = line.strip().split()
            llave = line[0]
            docid_path = line[1].split('/')
            docid = docid_path[-1]
            if(llave in list(resultados.keys())):
                resultados[llave].append(docid)
            else:
                resultados[llave] = [docid]
    avg_p_list=[]
    avg_r_list=[]
    avg_ndcg_list=[]
    for llave in list(resultados.keys()):
        rel_llave = llave
        if(llave[0]=='0'):
            rel_llave = llave[1:]
        relevantes = set(rels[rel_llave])
        precision_list = []
        recall_list = []
        rel_bool_list =[]
        for i in range(1,len(resultados[llave])+1):
            rel_doc = resultados[llave][i-1]
            recuperados = set(resultados[llave][:i])
            inter = relevantes & recuperados
            if(rel_doc in relevantes):
                precision = 100*float(len(inter))/i
                precision_list.append(precision)
                rel_bool_list.append(1)
            else:
                rel_bool_list.append(0)
            if(i == k):
                recall = 100*float(len(inter))/len(relevantes)
                recall_list.append(recall)
                break

        ndcg_query = ndcg_at_k(rel_bool_list, k,method=1)
        if(len(precision_list)!=0):
            avg_p = sum(precision_list)/len(precision_list)
        else:
            avg_p = 0.0
        if(len(recall_list)!=0):
            avg_r = sum(recall_list)/len(recall_list)
        else:
            avg_r = 0.0

        avg_p_list.append(avg_p)
        avg_r_list.append(avg_r)
        avg_ndcg_list.append(ndcg_query)
        if(verbose):
            print('***********************************')
            print("Query "+str(llave))
            print("@"+str(k))
            print("Avg. Precision: {:.4f}%".format(avg_p))
            print("Avg. Recall: {:.4f}%".format(avg_r))
            print("NDCG: {:.4f}%".format(ndcg_query))
    avg_p_modelo = sum(avg_p_list)/len(avg_p_list)
    avg_r_modelo = sum(avg_r_list)/len(avg_r_list)
    avg_ndcg_modelo = sum(avg_ndcg_list)/len(avg_ndcg_list)
    print('----------------------------------------------')
    print("TOTAL MODELO: "+files)
    print("@"+str(k))
    print("Avg. Precision: {:.4f}%".format(avg_p_modelo))
    print("Avg. Recall: {:.4f}%".format(avg_r_modelo))
    print("Avg. NDCG: {:.4f}%".format(avg_ndcg_modelo))
    return avg_p_list, avg_r_list, avg_ndcg_list
```
